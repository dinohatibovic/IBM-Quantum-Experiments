# Bell State Entanglement on IBM Quantum Hardware
This notebook reproduces the Bell state experiment executed on IBM Heron R2 and Heron R1.

**Metrics (from the verified Zenodo dataset, DOI 10.5281/zenodo.20749395):**
- Fidelity: 96.3% (Heron R2)
- Concurrence: 0.926
- Tangle: 0.858
- Backend comparison: +8.3 pp fidelity advantage


```python
from qiskit_ibm_runtime import QiskitRuntimeService, Sampler
from qiskit import QuantumCircuit
import numpy as np
import matplotlib.pyplot as plt

service = QiskitRuntimeService()
backend_r2 = service.backend('ibm_fez')
backend_r1 = service.backend('ibm_torino')
sampler_r2 = Sampler(backend=backend_r2)
sampler_r1 = Sampler(backend=backend_r1)
```

## Create Bell Circuit

```python
def bell_circuit():
    qc = QuantumCircuit(2)
    qc.h(0)
    qc.cx(0, 1)
    qc.measure_all()
    return qc

qc = bell_circuit()
```

## Execute on Heron R2 and Heron R1

```python
result_r2 = sampler_r2.run(qc, shots=1000).result()
result_r1 = sampler_r1.run(qc, shots=1000).result()

dist_r2 = result_r2.quasi_dists[0]
dist_r1 = result_r1.quasi_dists[0]
```

## Plot Probability Distributions

```python
plt.bar(dist_r2.keys(), dist_r2.values(), label='Heron R2')
plt.bar(dist_r1.keys(), dist_r1.values(), alpha=0.5, label='Heron R1')
plt.legend()
plt.title('Bell State Probability Distribution')
plt.show()
```
